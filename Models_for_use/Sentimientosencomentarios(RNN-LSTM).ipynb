```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.text import Tokenizer
import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset

# ---------------------------------------------------------------------------------------------------

# Cargar el dataset de Hugging Face
dataset = load_dataset("dair-ai/emotion")

# Mapa de emociones:
# ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
#      0        1       2       3       4          5
indexOfSentiment = 3

# Crear un conjunto de datos binario para la emoción concreta
def binary_label(example):
    example['label'] = 1 if example['label'] == indexOfSentiment else 0
    return example

train_dataset = dataset['train'].map(binary_label)

train_texts = train_dataset['text']

# Tokenización y Padding
max_features = 10000
max_len = 100

tok = Tokenizer(num_words=max_features)
tok.fit_on_texts(train_texts)

# ---------------------------------------------------------------------------------------------------

# # Base de datos

# operation Github
def load_data():
    
    # Models
    modelLoadedJoblibSurprise = load_model("C:/Users/arian/OneDrive/Documentos/GitHub/Artificial-Intelligence-Resources-New/Download_Files/modelsUse/arianMind/surpriseModelCritic.keras")
    modelLoadedJoblibSad = load_model('C:/Users/arian/OneDrive/Documentos/GitHub/Artificial-Intelligence-Resources-New/Download_Files/modelsUse/arianMind/sadnessModelCritic.keras')
    modelLoadedJoblibFear = load_model('C:/Users/arian/OneDrive/Documentos/GitHub/Artificial-Intelligence-Resources-New/Download_Files/modelsUse/arianMind/fearModelCritic.keras')
    modelLoadedJoblibJoy = load_model('C:/Users/arian/OneDrive/Documentos/GitHub/Artificial-Intelligence-Resources-New/Download_Files/modelsUse/arianMind/joyModelCritic.keras')
    modelLoadedJoblibLove = load_model('C:/Users/arian/OneDrive/Documentos/GitHub/Artificial-Intelligence-Resources-New/Download_Files/modelsUse/arianMind/loveModelCritic.keras')
    modelLoadedJoblibAnger = load_model('C:/Users/arian/OneDrive/Documentos/GitHub/Artificial-Intelligence-Resources-New/Download_Files/modelsUse/arianMind/angerModelCritic.keras')

    # Return    
    return modelLoadedJoblibSad, modelLoadedJoblibAnger, modelLoadedJoblibFear, modelLoadedJoblibJoy, modelLoadedJoblibLove, modelLoadedJoblibSurprise

modelLoadedJoblibSad, modelLoadedJoblibAnger, modelLoadedJoblibFear, modelLoadedJoblibJoy, modelLoadedJoblibLove, modelLoadedJoblibSurprise = load_data()

# Función para predecir emociones
def preprocess_text(text):
    sequence = tok.texts_to_sequences([text])
    padded_sequence = pad_sequences(sequence, maxlen=max_len)
    return padded_sequence



```

```python
# Predecir emoción
# emotionMap = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
def predict_emotion(text):
    processed_text = preprocess_text(text)

    # Predicción modelos
    predictionSad = modelLoadedJoblibSad.predict(processed_text,verbose=0)
    predictionAnger = modelLoadedJoblibAnger.predict(processed_text,verbose=0)
    predictionFear = modelLoadedJoblibFear.predict(processed_text,verbose=0)
    predictionJoy = modelLoadedJoblibJoy.predict(processed_text,verbose=0)
    predictionLove = modelLoadedJoblibLove.predict(processed_text,verbose=0)
    predictionSurprise = modelLoadedJoblibSurprise.predict(processed_text,verbose=0)

    allEmotionPrediction = [predictionSad[0][0],predictionJoy[0][0],predictionLove[0][0],predictionAnger[0][0],predictionFear[0][0],predictionSurprise[0][0]]
    
    # Probabilidad de cada uno
    bestEmotion = max(allEmotionPrediction)
    guideEmotionMap = []

    for i in allEmotionPrediction:
        if bestEmotion == i:
            guideEmotionMap.append(allEmotionPrediction.index(i))

    # All results
    return guideEmotionMap

# Example data

# Mapa de emociones:
# ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
#      0        1       2       3       4          5

# # User variables
acciertosEmocionales = 0
validationDatasete = dataset['validation']['label']
validationTextete = dataset['validation']['text']
for z in range(0,len(dataset['validation']['text'])):
    userTextPrediction = predict_emotion(validationTextete[z])

    if validationDatasete[z] in userTextPrediction:
        acciertosEmocionales += 1

    print(f"Validación número: {z}")

print(f'predicción final: {(acciertosEmocionales/len(validationDatasete)*100)}%')
```
